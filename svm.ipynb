```python
clf = SVC(kernel="linear")
clf.fit(X,y)
support_vectors = clf.support_vectors_
number_of_support_vectors = len(support_vectors)
print("Sono stati imparati ",number_of_support_vectors," support vectors per il classificatore lineare di base")
print("Questi sono i support vectors: \n",support_vectors)
y_color = ["red" if x + y > 0.3 else "green"  for [x,y] in support_vectors]
plt.title("Support vectors classificatore di base")

plt.scatter(X[:,0], X[:,1], c=y)
plt.scatter(support_vectors[:,0], support_vectors[:,1], marker="o",s=50, color='b')
plt.ylim(-1.0,1.0)
plt.xlim(-1.0,1.0)
plt.show()

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
# plot support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
#cerco i margin errors, punti che hanno una distanza inferiore a uno dal decision boundary
distance_from_boundary = clf.decision_function(X)
margin_error_point_index = np.where(abs(distance_from_boundary)<1)
margin_error_point = [X[index] for index in margin_error_point_index]
ax.scatter(margin_error_point[:][0][:,0], margin_error_point[:][0][:,1], marker="o",s=50, color='b')
plt.title("Decision boundary with margins and margins error")
plt.show()
```
